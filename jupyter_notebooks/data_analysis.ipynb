### 查看数据

```python
def print_ex(data):
    for i in range(len(data)):
        item = data.iloc[i]
        s = str(item.answer)
        if (len(s) > 10 and len(s) < 100) or len(s) > 800:
            print(i, 'Q:', str(item.question), '||', 'A:', s)
```

```python
print_ex(data)
```
